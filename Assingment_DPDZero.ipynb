## Validation:

##  Ensure call_date, agent_id, and org_id are present and correctly formatted

```python
import logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
# Function to validate a dataframe
def validate_dataframe(df, df_name, required_columns):
    # 1. Check if required columns are present
    for col in required_columns:
        if col not in df.columns:
            logging.error(f"❌ Column '{col}' is missing in {df_name}")
        else:
            logging.info(f"✅ Column '{col}' exists in {df_name}")
# Task 1: Validate required columns for each dataset
validate_dataframe(call_logs, "call_logs", ["agent_id", "org_id", "call_date"])
validate_dataframe(agent_roster, "agent_roster", ["agent_id", "org_id"])
validate_dataframe(disposition_summary, "disposition_summary", ["agent_id", "org_id", "call_date"])
```

### Flag missing or duplicate entries

```python
def missing_value(df, df_name, required_columns):
    missing = df[required_columns].isnull().sum()
    for col, count in missing.items():
        if count > 0:
            logging.warning(f"⚠️ {count} missing values in '{col}' of {df_name}")
        else:
            logging.info(f"✅ No missing values in '{col}' of {df_name}")

    dup_count = df.duplicated().sum()
    if dup_count > 0:
        logging.warning(f"⚠️ {dup_count} duplicate rows found in {df_name}")
    else:
        logging.info(f"✅ No duplicate rows in {df_name}")

# Run the validation
missing_value(call_logs, "call_logs", ["agent_id", "org_id", "call_date"])
missing_value(agent_roster, "agent_roster", ["agent_id", "org_id"])
missing_value(disposition_summary, "disposition_summary", ["agent_id", "org_id", "call_date"])
```

# Task-2 

```python
# Step 2: Merge call_logs with disposition_summary on common keys
merged_step1 = pd.merge(call_logs, disposition_summary,
                        on=["agent_id", "org_id", "call_date"],
                        how="outer", indicator=True)

print("\n▶️ Merged DataFrame (Step 1):")
display(merged_step1) 

print("\n▶️ Merge Origin Summary:")
display(merged_step1["_merge"].value_counts().to_frame(name="count"))
```

```python
# Step 3: Merge result with agent_roster (based on agent_id, org_id only)
final_merged = pd.merge(merged_step1, agent_roster,
                        on=["agent_id", "org_id"],
                        how="outer", indicator="merge_status",
                        suffixes=('', '_roster'))

print("\n✅ Final merge with agent_roster complete:")
print("\n▶️ Final Merged DataFrame:")
display(final_merged.head()) 

print("\n▶️ Merge Status Summary:")
display(final_merged["merge_status"].value_counts().to_frame(name="count"))
display(final_merged.shape)
```

#### A full outer join is used to retain all records from all datasets, even if they don’t match.

```python
# Step 4: Identify unmatched rows
unmatched = final_merged[final_merged["merge_status"] != "both"]

print("\n⚠️ Unmatched Rows (mismatches across datasets):")
display(unmatched.head())  
print(f"\n🔎 Total unmatched rows: {len(unmatched)}")
```

# Task -3 

## Feature Engineering

```python
# Step 1: Create presence flag (1 if login_time exists, else 0)
final_merged["presence"] = final_merged["login_time"].notnull().astype(int)

# Step 2: Create a 'connected_flag' column for completed calls
final_merged["is_connected"] = final_merged["status"].str.lower().eq("completed").astype(int)
```

```python
# Aggregate key metrics per agent per day
agent_daily_summary = final_merged.groupby(["agent_id", "call_date"]).agg(
    total_calls=("call_id", "count"),
    unique_loans_contacted=("installment_id", "nunique"),
    completed_calls=("is_connected", "sum"),
    avg_call_duration_min=("duration", lambda x: round(x.mean(), 2)),
    presence=("presence", "max")
).reset_index()

# Add connect rate
agent_daily_summary["connect_rate"] = round(
    agent_daily_summary["completed_calls"] / agent_daily_summary["total_calls"], 2)
```

```python
print("🧾 Final Output (agent_daily_summary):")
display(agent_daily_summary)
```

## Format a Slack-style summary message like:

```python
# Generate summary for specific date (e.g., 2025-04-28)
report_date = "2025-04-28"
summary_df = agent_daily_summary[agent_daily_summary["call_date"] == report_date]

if not summary_df.empty:
    top_row = summary_df.loc[summary_df["connect_rate"].idxmax()]
    top_agent_name = final_merged.loc[final_merged["agent_id"] == top_row["agent_id"], "users_first_name"].iloc[0]
    top_connect_rate = int(top_row["connect_rate"] * 100)
    active_agents = summary_df["agent_id"].nunique()
    avg_duration = round(summary_df["avg_call_duration_min"].mean(), 3)
    
    slack_message = (
        f"*Agent Summary for {report_date}*\n"
        f"Top Performer: {top_agent_name} ({top_connect_rate}% connect rate)\n"
        f"Total Active Agents: {active_agents}\n"
        f"Average Duration: {avg_duration} min"
    )
else:
    slack_message = f"*Agent Summary for {report_date}*\nNo activity recorded."

print("\n📩 Slack Summary Message:")
print(slack_message)
```
